### Motion in Straight Line:
### s, v, a Dependent on t \[s(t), v(t), a(t)\]:

$$\large v(t) = \frac{ds(t)}{dt} \:,\: a(t) = \frac{dv(t)}{dt} $$

$$\large \int_{s(t_0)=s_0}^{s(t_f)=s_f}{ds} = \int_{t_0}^{t_f}{v(t) \: dt}$$

$$\large \int_{v(t_0)= v_0}^{v(t)=v}{dv} = \int_{t_0}^{t}{a(t) \: dt}$$

### a Dependent on v \[a(v)\]:

Known data relates time (t) and velocity (v).

$$\large\int_{t_0}^{t}{dt} = \int_{v(t=t_0) = v_0}^{v(t) = v}{\frac{dv}{a(v)}} = \int_{v_0}^{v}{\frac{dv}{a(v)}}$$

Known data relates velocity (v) and position (s).

$$\large \int_{s_0}^{s}{ds} = \int_{v(s=s_0) = v_0}^{v(s) = v}{\frac{v \: dv}{a(v)}} = \int_{v_0}^{v}{\frac{v \: dv}{a(v)}}$$

### a and/or v Dependent on s \[a(s), v(s)\]:

$$\large a(s) = \frac{dv}{ds}\: v(s) $$

$$\large \int_{s_0}^{s}{a(s) \: ds} = \int_{v_0}^{v}{v \: dv}$$

$$\large \int_{t_0}^{t}{dt} = \int_{s_0}^{s}{\frac{ds}{v(s)}}$$

### Position, Velocity, Acceleration
#### 13.1

#### Position: **$\vec{r}$**  

In this class we use a generalized **$\vec{r}$** vector to describe the position of a point like object. The thing to notice is that the use of **$\vec{r}$** does **NOT** imply that we are working in spherical coordinates as we might if we are in math class. In addition **$\vec{r}$** may depend on any one of a number of variables. This is a step in a new direction from our previous classes where we always parameterized the position using time (t). It is certainly true that we will often know **$\vec{r}$** as a function of time but don't be misled into thinking that this is universal.

#### Velocity: $\vec{v} = \frac{d\vec{r}}{dt}$

Regardless of what the dependence of **$\vec{r}$** is velocity is defined as the rate of change with respect to time. How this is accomplished can involve product rules, chain rules, or implicit differentiation using tools from MTH 251. As an example:

$$\large \frac{d\vec{r}(s)}{dt} = \frac{d\vec{r}(s)}{ds}\frac{ds}{dt}$$

Because **$\vec{r}$** is a vector we may well have to consider the rate of change of the unit vectors embedded in **$\vec{r}$** along the way. We'll get to this a bit later.

It is important to note that because $\vec{\Delta r} = \vec{r_2} - \vec{r_1}$ we need not be concerned with where we pick the origin of our coordinate system or even which coordinate system we use. The differences between 2 vectors are independent of the choice of coordinate system. This means neither the velocity nor the acceleration depend on the coordinate system. 

**Note:** Later we will consider dynamic behavior in different coordinate systems where we will see different velocity and acceleration terms depending on the choice coordinate system. This is not a violation of what I just stated but rather a consequence of seeking to apply Newton's Laws in a coordinate system that doesn't not satisfy Newton's First Law. We'll talk more about this when we get to that point.

#### Acceleration: $\vec{a} = \frac{d^2\vec{r}}{dt^2} = \frac{d\vec{v}}{dt}$

Nothing new here but all the same caveats apply that I noted in the discussion of the velocity.

### Straight Line Motion:
#### 13.2

Let us assume for now that we know the path of an object through the universe. Like the red path shown in the plot above. 's' is a standard symbol to label where we are along that path. 's' is a location and is measured in the usual units of distance. We can define a unit vector tangent to the path along which the particle is traveling everywhere along this path that we might call $\vec{e_s}$. In the particular case of straight line motion **where the origin is on the line of motion** then $\vec{r} = s\vec{e_s}$ where $\vec{e_s}$ never changes direction. In the plot below you can see that the blue vectors (barely visible on top of the red path) satisfy this relationship while the green ones do not.

<img src="https://raw.githubusercontent.com/smithrockmaker/ENGR212/master/images/straightline.png" />

Given

$$\large \vec{r} = s\vec{e_s}$$

then

$$\large \vec{v} = \frac{d\vec{r}}{dt} = \frac{ds}{dt}\vec{e_s} + s\frac{d\vec{e_s}}{dt}$$

for straight line motion

$$\large \frac{d\vec{e_s}}{dt} = 0 $$

but hopefully you notice that this will not always be the case in general. Under these circumstances

$$\large \vec{v} = \frac{ds}{dt}\vec{e_s} \implies v =  \frac{ds}{dt}$$ 

Similarly 

$$\large \vec{a} = \frac{d\vec{v}}{dt} = \frac{d}{dt}(\frac{ds}{dt})\vec{e_s} + v\frac{d\vec{e_s}}{dt}$$

Leading us to (for this special case of straight line motion where $ \frac{d\vec{e_s}}{dt} = 0 $)

$$\large \vec{a} = \frac{d^2s}{dt^2}\vec{e_s} = \frac{dv}{dt}\vec{e_s} $$


### s(t), v(t), or a(t) Known:

**Note:** Still for straight line motion

If any one of s(t), v(t), or a(t) are known functions then one can determine the others through either differentiation or integration. The relationship among the derivatives is described in the previous cell. While we have solved integration problems of this type in math class we will take a slightly different approach here that explicitly addresses the boundary conditions of a particular problem setting. 

#### v(t) Known:

"v(t) Known" means that you have been the actual mathematical function of t that describes v(t) for every value of t in some range. Given $ v(t) = \frac{ds}{dt} $ and the standard incantation of the applied sciences ("For smooth and well behaved functions - i.e. real world behavior") we can treat the derivative as a fraction and rewrite this as $ v(t)\: dt = ds $. Integrating both sides between the same points in the process we get..

$$\large \int_{s(t_0)=s_0}^{s(t_f)=s_f}{ds} = \int_{t_0}^{t_f}{v(t) \: dt}$$

It is important conceptually to be confident that you understand how the limits on both sides of the expression are related to each other. We are not just tossing limits onto these integrals. What is on the left side of the equation must be describing the same portion of the progress of the particle as the right side.

#### a(t) Known:

If a(t) is known explicitly, in the same sense as our previous discussion of v(t) being known, then we can make an identical argument.

$$\large a(t) = \frac{dv}{dt} \implies a(t)dt = dv $$

leading to

$$\large \int_{v(t_0)= v_0}^{v(t)=v}{dv} = \int_{t_0}^{t}{a(t) \: dt}$$

The results of this will give, after some algebra, an explict function v(t) which we can integrate again to find s(t).

#### a(t) = constant = $a_c$:

This is a special case that we explored in detail during PH211 and you are aware that our 'kinematic equations' from that class are the result of doing the integrals we have just discussed. Those kinematics equations are:

.$$\large \label{eqn1}\tag{1} v(t) = v_0 + a_c (t - t_0) $$

.$$\large \label{eqn2}\tag{2} s(t) = s_0 + v_0 (t - t_0) + \frac{1}{2} a_c (t - t_0)^2 $$

...and if we solve (\ref{eqn1}) for $(t- t_0)$ and substitute it into  (\ref{eqn2}) we get...

$$\large \label{eqn3}\tag{3} v(t)^2 = v_0^2 + 2 \: a_c (s - s_0) $$

### a(v) is Known:
#### 13.3

So what if we don't know the explicit time dependence of a dynamic variable. What if, for instance, we know the dependence of acceleration on velocity? Consider the basic definition of acceleration...

$$\large \label{eqn4}\tag{4} a(v) = \frac{dv}{dt} $$

After muttering the usual incantation we can rearrange this to give...

$$\large dt = \frac{dv}{a(v)} $$

....which, if we know the explicit functional dependence of a(v), can be integrated...

$$\large \label{eqn5}\tag{5} \int_{t_0}^{t}{dt} = \int_{v(t=t_0) = v_0}^{v(t) = v}{\frac{dv}{a(v)}} = \int_{v_0}^{v}{\frac{dv}{a(v)}}$$

The right hand side of this equation will contain v, possibly in a complex function, and the left hand side will be t. In principle that means that mathematically we can now determine v(t). Once we have v(t) we can use our previous approach to find s(t). While this may be possible it can easily be quite complex. Returning to (\ref{eqn4}) we can use the chain rule to rewrite it as....

$$\large \label{eqn6}\tag{6} a(v) = \frac{dv}{dt} = \frac{dv}{ds}\frac{ds}{dt} = \frac{dv}{ds}\: v $$

Rearranging again we get...

$$\large \label{eqn7}\tag{7} ds = \frac{v \: dv}{a(v)}$$

..and integrating....

$$\large \label{eqn8}\tag{8} \int_{s_0}^{s}{ds} = \int_{v_0}^{v}{\frac{v \: dv}{a(v)}}$$

This is a lot to keep track of and we will summarize an approach to problem solving in a moment.

### a(s) or v(s) is Known:

We follow a similar process of if we know the explicit dependence of a() on s. Going back to (\ref{eqn6}) we can write...

$$\large a(s) = \frac{dv}{dt} = \frac{dv}{ds}\frac{ds}{dt} = \frac{dv}{ds}\: v $$

..which, after rearranging, gives...

$$\large \label{eqn9}\tag{9} a(s) \: ds = v \: dv $$

..which we can integrate...

$$\large \label{eqn10}\tag{10} \int_{s_0}^{s}{a(s) \: ds} = \int_{v_0}^{v}{v \: dv}$$

This gives us an expression for v(s).

$$\large \label{eqn11}\tag{11} v(s) = \frac{ds}{dt} \implies dt = \frac{ds}{v(s)}$$

..which we integrate..

$$\large \label{eqn12}\tag{12} \int_{t_0}^{t}{dt} = \int_{s_0}^{s}{\frac{ds}{v(s)}}$$

This gives us the position as a function of time.


### Problem Solving Choices:

None of the individual parts of the discussion above tend to bother students. It all makes sense based on your understanding of calculus. The challenge comes in trying to identify which approach to take under what circumstances. Your first step after setting up the problem is to identify the independent variable in the problem. Do you have a known function of t, v, or s?

#### a(t), v(t), or s(t) Known: Dependence on t

From a practical point of view this is the sort of data you might take from a video clip or radar data.

Use the standard definitions of v(t) and a(t) and differentiate or integrate with respect to time to find the desired dependence on time. Start with (\ref{eqn13}) and separate variable if you need to integrate.

$$\large \label{eqn13}\tag{13} v(t) = \frac{ds(t)}{dt} \:,\: a(t) = \frac{dv(t)}{dt} $$

#### a(v) Known: Dependence on v

From a practical perspective this is the sort of data you might have if you are determining accelerations from velocity dependent forces like drag forces.

Use the following to definitions of a(v) to separate variables and integrate to solve for s and t as functions of v.

$$\large \label{eqn14}\tag{14} a(v) = \frac{dv}{dt} \: , \: a(v) = \frac{dv}{ds}\frac{ds}{dt} = \frac{dv}{ds}\: v $$

#### a(s) or v(s) Known: Dependence on s

When we have forces that are location dependent like gravity or EM forces then we can end up with position dependent accelerations. Forces in expanding chambers are also position dependent. 

Use the following definitions of a(s) and v(s) to separate variables and integrate to solve for v and t as functions of s.

$$\large \label{eqn15}\tag{15} a(s) = \frac{dv}{ds}\frac{ds}{dt} = \frac{dv}{ds}\: v \:, \: v(s) = \frac{ds}{dt}$$
